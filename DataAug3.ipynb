```
import os
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import numpy as np
from imutils import paths
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from keras.optimizers import SGD
from keras.utils import np_utils
import cv2
from google.colab import drive, files
```

```
train_data = np.load('training.npy')
target_data = np.load('testing.npy')
```

```
print("There are {} images in the training dataset".format(len(train_data)))
print('There are {} image targets in testing dataset'.format(len(target_data)))
print('There are {} unique targets in the testing dataset'.format(len(np.unique(target_data))))
print("Size of each image is {}x{}".format(train_data.shape[1],train_data.shape[2]))
```

```
plt.imshow(train_data[0])
```

```
X=train_data.reshape((train_data.shape[0],train_data.shape[1]*train_data.shape[2]*train_data.shape[3]))
print("X shape:",X.shape)
```

```
X_train, X_test, y_train, y_test=train_test_split(X, target_data, test_size=0.3, stratify=target_data, random_state=0)
print("X_train shape:",X_train.shape)
print("y_train shape:{}".format(y_train.shape))
```

```
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
clf = SVC()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("accuracy score:{:.2f}".format(accuracy_score(y_test, y_pred)))
```

```
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("accuracy score:{:.2f}".format(accuracy_score(y_test, y_pred)))
```

```
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("accuracy score:{:.2f}".format(accuracy_score(y_test, y_pred)))
```
